```python
import pandas as pd
import numpy as np
df = pd.read_csv("./eplmatches.csv")
```

```python
season_needed = df[df.Season_End_Year == 2007]
```

```python
home = season_needed[["Home", "HomeGoals", "AwayGoals", "FTR"]]
home["FTR"] = home.FTR.apply(lambda x: {'H': 'W', 'A': 'L', 'D': 'D'}[x])
away = season_needed[["Away", "AwayGoals", "HomeGoals", "FTR"]]
away["FTR"] = away.FTR.apply(lambda x: {'A': 'W', 'H': 'L', 'D': 'D'}[x])
```

```python
home.columns = ['Team', 'BeatenGoals', 'MissedGoals', 'Wins']
away.columns = ['Team', 'BeatenGoals', 'MissedGoals', 'Wins']
```

```python
common = pd.concat([home, away])
```

```python
teams = common.groupby('Team')
wr = teams[["Wins"]].value_counts()
```

```python
joined = teams.agg(sum)
```

```python
w = wr.loc[common.Team, "W"]
d = wr.loc[common.Team, "D"]
l = wr.loc[common.Team, "L"]
```

```python
stats = pd.DataFrame({'Wins': w, 'Draw': d, 'Loses': l}).groupby('Team').sum()
```

```python
res = joined.merge(stats, on='Team')
res.columns=['BeatenGoals', 'MissedGoals', 'Win', 'Draw', 'Lose']
res['Points'] = res.Win * 3 + res.Draw
```

```python
good_res = res[["Points", "Win", "Draw", "Lose", "BeatenGoals", "MissedGoals"]].astype('int')
good_res['Help'] = good_res.BeatenGoals - good_res.MissedGoals
good_res = good_res.sort_values(by=['Points', 'Help', 'Win'], ascending=False)
good_res = good_res.drop('Help', axis=1)
```
